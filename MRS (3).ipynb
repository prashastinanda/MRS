```python
import ast
def convert(obj):
    l=[];
    for i in ast.literal_eval(obj):
        #String of list need to converted into list to apply the function
        l.append(i['name'])
    return l
```

```python
# we need to extract top 3 actors name from cast
def convert3(obj):
    l=[];
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
# need to fetch only director name
import ast
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
def recommend(movie):
    if movie not in new_df['title'].values:
        return "Movie not found in dataset!"

    movie_index = new_df[new_df['title'] == movie].index[0]
    similarity_scores = list(enumerate(cosine_sim[movie_index]))
    sorted_movies = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:11]  # Top 10 similar movies

    recommended_movies = []
    for i in sorted_movies:
        recommended_movies.append(new_df.iloc[i[0]].title)

    return recommended_movies
```

```python
#collaborative filtering

# Compute the mean vote across the dataset
C = new_df['vote_average'].mean()

# Compute the 90th percentile of vote count
m = new_df['vote_count'].quantile(0.90)
```

```python
# Compute weighted rating (IMDB formula) without filtering
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m) * R) + (m / (v + m) * C)

# Apply weighted rating to all movies
new_df['score'] = new_df.apply(weighted_rating, axis=1)

# Sort movies based on score
new_df = new_df.sort_values('score', ascending=False)
```

```python
from scipy.sparse.linalg import svds

# Select relevant columns for matrix factorization
new_df = new_df[['id', 'title', 'vote_average', 'vote_count', 'score']]
```

```python
from scipy.sparse import csr_matrix
# Normalize vote_count
new_df['normalized_votes'] = (new_df['vote_count'] - new_df['vote_count'].min()) / \
                             (new_df['vote_count'].max() - new_df['vote_count'].min())

# 🔹 Ensure unique movie_id by keeping the highest vote count movie
new_df = new_df.sort_values(by='vote_count', ascending=False).drop_duplicates(subset=['id'])

# 🔹 Create user-item matrix with `vote_average` (ratings)
movie_matrix = new_df.pivot(index='id', columns='title', values='vote_average').fillna(0)

# 🔹 Convert DataFrame to sparse matrix (better for SVD)
movie_matrix_sparse = csr_matrix(movie_matrix.values.astype(np.float64))

# 🔹 Apply Singular Value Decomposition (SVD)
U, sigma, Vt = svds(movie_matrix_sparse, k=min(10, min(movie_matrix.shape)-1))  # Ensure k is valid

# Convert sigma into a diagonal matrix
sigma = np.diag(sigma)

# Compute predicted "ratings"
predicted_ratings = np.dot(np.dot(U, sigma), Vt)

# 🔹 Convert predictions back into a DataFrame
predicted_df = pd.DataFrame(predicted_ratings, index=movie_matrix.index, columns=movie_matrix.columns)
```

```python
def recommend_movies(movie_title, num_recommendations=10):
    if movie_title not in predicted_df.columns:
        return "Movie not found in dataset!"

    # Fetch similarity scores for the given movie
    movie_vector = predicted_df[movie_title]

    # Compute similarity scores
    similarity_scores = movie_vector.dot(predicted_df)

    # Sort movies by similarity score
    similar_movies = similarity_scores.sort_values(ascending=False).iloc[1:num_recommendations+1]

    return similar_movies.index.tolist()

# --- Example Usage ---
print(recommend_movies('Avatar', num_recommendations=10))  # Replace 'Avatar' with any movie title
```

```python
# Function to normalize scores
def normalize_scores(scores):
    return (scores - np.min(scores)) / (np.max(scores) - np.min(scores)) if np.max(scores) > np.min(scores) else scores
```

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import mean_squared_error

# Load the movies dataset
movies_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/movies_metadata.csv")

# Ensure 'id' column is numeric
movies_df['id'] = pd.to_numeric(movies_df['id'], errors='coerce')
movies_df = movies_df.dropna(subset=['id'])
movies_df['id'] = movies_df['id'].astype(int)

# Generate synthetic ratings data
np.random.seed(42)
ratings_data = {
    'userId': np.random.randint(1, 100, size=2000),
    'movieId': np.random.choice(movies_df['id'].unique(), size=2000, replace=True),
    'rating': np.random.randint(1, 6, size=2000)
}
ratings_df = pd.DataFrame(ratings_data)

# Filter out users with less than 10 ratings
user_counts = ratings_df['userId'].value_counts()
valid_users = user_counts[user_counts > 10].index
ratings_df = ratings_df[ratings_df['userId'].isin(valid_users)]

# Create user-item matrix
user_item_matrix = ratings_df.pivot_table(index='userId', columns='movieId', values='rating').fillna(0)

# Normalize ratings (subtract user mean)
user_means = user_item_matrix.mean(axis=1)
user_item_matrix_centered = user_item_matrix.sub(user_means, axis=0).fillna(0)

# Apply Truncated SVD with increased latent dimensions
svd = TruncatedSVD(n_components=75, random_state=42)
reduced_matrix = svd.fit_transform(user_item_matrix_centered)
reconstructed_matrix = np.dot(reduced_matrix, svd.components_)

# Add user means back
reconstructed_matrix += user_means.values[:, np.newaxis]

# Calculate RMSE
mask = user_item_matrix.values > 0
rmse = np.sqrt(mean_squared_error(user_item_matrix.values[mask], reconstructed_matrix[mask]))

# Output the improved RMSE
print(f"Improved RMSE (SVD): {rmse:.4f}")
```

```python
# Hybrid filtering: Weighted combination of both methods
def hybrid_recommendations(movie_title, content_weight=0.5, collab_weight=0.5, top_n=10):
    movie_idx = new_df.index[new_df['title'] == movie_title].tolist()[0]
    content_scores = content_similarity[movie_idx]

    collab_scores = predicted_df.loc[movie_title] if movie_title in predicted_df.index else np.zeros(len(content_scores))

    # Normalize scores
    content_scores = normalize_scores(content_scores)
    collab_scores = normalize_scores(collab_scores)

    # Hybrid score
    hybrid_scores = content_weight * content_scores + collab_weight * collab_scores

    # Get top N recommendations
    recommended_indices = np.argsort(hybrid_scores)[::-1][1:top_n+1]
    return new_df.iloc[recommended_indices]['title']
```
